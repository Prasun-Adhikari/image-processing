```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_o = plt.imread('nasa2.png')
img = img_o[:,:,0]
M, N = img.shape
```

```python
plt.imshow(img, cmap='grey')
```

```python
f = img / 256
t = np.fft.fft(f)
f2 = np.fft.ifft(t)
```

```python
mask = np.zeros((M, N))
d = 25
mask[:,d] = 0.05 - np.sin(np.arange(M) * 0.275 + (-1/4) * np.pi) * 0.2
mask[:,N-d] = 0.05 - np.sin(np.arange(M) * 0.275 + (-1/4) * np.pi) * 0.2
f3 = np.fft.ifft(t - mask)
plt.imshow(abs(t - mask))
plt.imshow(abs(f3), cmap='grey')
```
