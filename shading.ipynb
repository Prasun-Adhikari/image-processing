```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d
```

```python
img_in = plt.imread("shaded_grid.png")
img0 = img_in[:,:,0]
M, N = img0.shape
```

```python
mk, nk = M//4, N//4
std = M // 16
```

```python
plt.imshow(img0, cmap='grey')
```

```python
gaussian_kernel = np.zeros((mk, nk))
xc, yc = (mk-1)//2, (nk-1)//2
for x in range(mk):
    for y in range(nk):
        r2 = (x-xc)**2 + (y-yc)**2
        gaussian_kernel[x,y] = np.exp(-r2 / (2 * std**2))
```

```python
mk, nk
```

```python
img1 = convolve2d(img0, gaussian_kernel, mode="same", boundary="symm") / np.sum(gaussian_kernel)
```

```python
plt.imshow(img1, cmap='grey')
```

```python
img2 = img0 / img1
plt.imshow(img2, cmap='grey')
```
